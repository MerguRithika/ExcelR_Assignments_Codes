```python
# Remove outliers for all numeric columns
for col in numeric_col:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR
    
    df = df[(df[col] >= lower_limit) & (df[col] <= upper_limit)]
```

```python
Y = df["Type"]
```

```python
X = df.drop(df[['Type']],axis=1)
X
```

```python
from sklearn.preprocessing import StandardScaler
SS = StandardScaler()
SS_X = SS.fit_transform(X)
SS_X = pd.DataFrame(SS_X)
SS_X.columns = list(X)
SS_X.head()
```

```python
from sklearn.decomposition import PCA

# Apply PCA (initially without fixing number of components)
pca = PCA()
X_pca = pca.fit_transform(SS_X)
```

```python
X_pca
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(8,5))
plt.plot(np.cumsum(pca.explained_variance_ratio_), marker='o')
plt.xlabel("Number of Principal Components")
plt.ylabel("Cumulative Explained Variance")
plt.title("Scree Plot / Cumulative Explained Variance")
plt.grid(True)
plt.show()
```

```python
optimal_components = np.argmax(np.cumsum(pca.explained_variance_ratio_) >= 0.95) + 1 
print("Optimal number of components:", optimal_components)
```

```python
pca_final = PCA(n_components=optimal_components)
X_pca_final = pca_final.fit_transform(SS_X)
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score

kmeans_original = KMeans(n_clusters=3, random_state=42)
labels_original = kmeans_original.fit_predict(SS_X)


def plot_clusters(X, labels, title="Clustering Results", cmap="viridis"):
    """
    Plots clustering results using the first two dimensions of X.
    Works for both NumPy arrays and Pandas DataFrames.
    """
  
    if hasattr(X, "iloc"):   
        X_plot = X.values
    else:                   
        X_plot = X
    
    plt.figure(figsize=(8,6))
    plt.scatter(X_plot[:, 0], X_plot[:, 1], c=labels, cmap=cmap, s=50)
    plt.title(title)
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.show()

plot_clusters(SS_X, labels_original, title="Clustering on Original Data")


sil_original = silhouette_score(SS_X, labels_original)
db_original = davies_bouldin_score(SS_X, labels_original)

print("Original Data - Silhouette Score:", sil_original)
print("Original Data - Davies-Bouldin Index:", db_original)
```

```python
kmeans_pca = KMeans(n_clusters=3, random_state=42)  
labels_pca = kmeans_pca.fit_predict(X_pca_final)


plot_clusters(X_pca_final, labels_pca, title="Clustering on PCA-Transformed Data", cmap="plasma")

sil_pca = silhouette_score(X_pca_final, labels_pca)
db_pca = davies_bouldin_score(X_pca_final, labels_pca)

print("PCA Data - Silhouette Score:", sil_pca)
print("PCA Data - Davies-Bouldin Index:", db_pca)

print("\n--- Comparison ---")
print("Original Data - Silhouette:", sil_original, " | Davies-Bouldin:", db_original)
print("PCA Data     - Silhouette:", sil_pca, " | Davies-Bouldin:", db_pca)

if sil_pca > sil_original:
    print(" PCA improved clustering performance.")
else:
    print("PCA did not improve clustering performance significantly.")

print("Trade-off: PCA reduces dimensionality (faster, less noise) but may lose some information.")
```

```python
# Task 5: Comparison and Analysis

import matplotlib.pyplot as plt

# Step 1: Compare clustering metrics
print("\n--- Clustering Performance Comparison ---")
print("Original Data - Silhouette Score:", sil_original, " | Davies-Bouldin Index:", db_original)
print("PCA Data     - Silhouette Score:", sil_pca, " | Davies-Bouldin Index:", db_pca)

# Step 2: Side-by-side visualization of clustering results
fig, axes = plt.subplots(1, 2, figsize=(14,6))

# Original Data Clustering
if hasattr(SS_X, "iloc"):
    X_plot_orig = SS_X.values
else:
    X_plot_orig = SS_X

axes[0].scatter(X_plot_orig[:,0], X_plot_orig[:,1], c=labels_original, cmap="viridis", s=50)
axes[0].set_title("Clustering on Original Data")
axes[0].set_xlabel("Feature 1")
axes[0].set_ylabel("Feature 2")

# PCA Data Clustering
if hasattr(X_pca_final, "iloc"):
    X_plot_pca = X_pca_final.values
else:
    X_plot_pca = X_pca_final

axes[1].scatter(X_plot_pca[:,0], X_plot_pca[:,1], c=labels_pca, cmap="plasma", s=50)
axes[1].set_title("Clustering on PCA Data")
axes[1].set_xlabel("PC1")
axes[1].set_ylabel("PC2")

plt.tight_layout()
plt.show()

# Step 3: Reflection in code output
if sil_pca > sil_original and db_pca < db_original:
    print("\nâœ… PCA improved clustering performance (better separation, lower DB index).")
elif sil_pca < sil_original and db_pca > db_original:
    print("\nâš ï¸ PCA reduced clustering performance (lost information during dimensionality reduction).")
else:
    print("\nâ„¹ï¸ PCA gave similar clustering performance compared to original data.")

# Step 4: Trade-off analysis (printed summary)
print("\n--- Trade-offs ---")
print("Original Data: Captures all feature interactions, but slower and harder to visualize.")
print("PCA Data: Faster, less noisy, easier to visualize, but may lose subtle information.")
```
