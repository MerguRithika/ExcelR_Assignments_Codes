```python
import pandas as pd 
df = pd.read_csv("Patient_Data.csv") # patient data
df2 = pd.read_csv("Billing_Data.csv") #Billing data
```

```python
# 1.	Load the patient dataset and show summary with info().
df.info()
df2.info()
```

```python
#2.	Select only the columns relevant for billing: ['PatientID', 'Department', 'Doctor', 'BillAmount'].
select = df[['PatientID', 'Department', 'Doctor', 'BillAmount']]
print(select)
```

```python
#3.	Drop administrative columns like ['ReceptionistID', 'CheckInTime'].
df.drop(['ReceptionistID', 'CheckInTime'], axis=1,inplace = True)
df.info()
```

```python
#4.	Use groupby to find total bill amount per department.
df.groupby("Department")['BillAmount'].sum().reset_index()
```

```python
# 5.Remove duplicate patient records based on PatientID.
df = df.drop_duplicates(subset="PatientID")
df.head()
```

```python
df.isnull().sum()
```

```python
mean_value = df["BillAmount"].mean()
```

```python
#6.	Fill missing BillAmount values with the mean bill amount
df['BillAmount'].fillna(value=mean_value,inplace=True)
print(df.head())
```

```python
df.isnull().sum()
```

```python
# 7.	Merge the billing dataset with patient dataset on PatientID.
merged_df = pd.merge(df2, df, on='PatientID', how='left')
print(merged_df.head())
```

```python
# 8.	Concatenate an additional DataFrame that contains new patients for the current week (row-wise).
new_patients = pd.DataFrame({
    'PatientID': [106,107],
    'Name': ['chiru', 'riya'],
    'Department': ['Cardiology', 'Orthopedics'],
    'Doctor': ['Dr. srija', 'Dr.pooja'],
    # Add more columns if your df2 has them
})

final_df= pd.concat([df2, new_patients], axis=0, ignore_index=True)
print(final_df.tail())
```

```python
# 9.	Concatenate new billing category columns like ['InsuranceCovered', 'FinalAmount'] (column-wise).
insurance_list = [1] * len(df2)
final_amount_list = df['BillAmount'] * 0.9

new_cols_df = pd.DataFrame({
    'InsuranceCovered': insurance_list,
    'FinalAmount': final_amount_list
})
df_with_new_cols = pd.concat([df, new_cols_df], axis=1)
print(df_with_new_cols.head())
```
