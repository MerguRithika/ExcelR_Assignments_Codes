```python
# DEFINING X AND Y VARIABLES
Y = df["Price"]
X = df.drop("Price",axis=1)
print(X.columns)
```

```python
# 2.Split the dataset into training and testing sets (e.g., 80% training, 20% testing).
# DATA PARTITION 
from sklearn.model_selection import train_test_split 
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.20, random_state = 42) # random_state 
```

```python
# checking whether data is partitioned or not
X_train.shape,X_test.shape,Y_train.shape,Y_test.shape
```

```python
# 3 Build a multiple linear regression model using the training dataset. Interpret the coefficients of the model. Build a minimum of 3 different models.
# 4.Evaluate the performance of the model using appropriate evaluation metrics on the testing dataset.
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error,r2_score

linreg1 = LinearRegression()
linreg1.fit(X_train, Y_train)
Y_pred1 = linreg1.predict(X_test)
print("\nMODEL 1: Using all available features")
coef_df1 = pd.DataFrame({
    "Feature": X.columns,
    "Coefficient": linreg1.coef_
})
print(coef_df1)
print("Intercept:", linreg1.intercept_)

# Evaluation for Model 1
mse1 = mean_squared_error(Y_test, Y_pred1)
rmse1 = np.sqrt(mse1)
r21 = r2_score(Y_test, Y_pred1)

print("\nModel 1 Performance:")
print("MSE :", mse1)
print("RMSE:", rmse1)
print("R^2 :", r21)





```

```python
#  MODEL 2: Only important numeric predictors
features_model2 = [c for c in ["Age", "KM", "HP", "Weight"] if c in X.columns]

X_train_m2 = X_train[features_model2]
X_test_m2 = X_test[features_model2]

linreg2 = LinearRegression()
linreg2.fit(X_train_m2, Y_train)

Y_pred2 = linreg2.predict(X_test_m2)

print("\nMODEL 2: Using selected numeric predictors:", features_model2)
coef_df2 = pd.DataFrame({
    "Feature": features_model2,
    "Coefficient": linreg2.coef_
})
print(coef_df2)
print("Intercept:", linreg2.intercept_)

# Evaluation for Model 2
mse2 = mean_squared_error(Y_test, Y_pred2)
rmse2 = np.sqrt(mse2)
r22 = r2_score(Y_test, Y_pred2)

print("\nModel 2 Performance:")

print("MSE :", mse2)
print("RMSE:", rmse2)
print("R^2 :", r22)
```

```python
#  MODEL 3: Drop potentially weak/collinear variables 
# Example: use all features except CC and Doors (just an assumption)
features_model3 = [col for col in X.columns if col not in ["CC", "Doors"]]

X_train_m3 = X_train[features_model3]
X_test_m3 = X_test[features_model3]

linreg3 = LinearRegression()
linreg3.fit(X_train_m3, Y_train)

Y_pred3 = linreg3.predict(X_test_m3)

print("\nMODEL 3: Dropping CC and Doors (if present)")
print("Features used:", features_model3)
coef_df3 = pd.DataFrame({
    "Feature": features_model3,
    "Coefficient": linreg3.coef_
})
print(coef_df3)
print("Intercept:", linreg3.intercept_)

# Evaluation for Model 3
mse3 = mean_squared_error(Y_test, Y_pred3)
rmse3 = np.sqrt(mse3)
r23 = r2_score(Y_test, Y_pred3)

print("\nModel 3 Performance:")

print("MSE :", mse3)
print("RMSE:", rmse3)
print("R^2 :", r23)

```

```python

# COMPARE MODEL PERFORMANCE

results = pd.DataFrame({
    "Model": ["Model 1 (All)", "Model 2 (Selected)", "Model 3 (Drop CC, Doors)"],
 
    "RMSE": [rmse1, rmse2, rmse3],
    "R2": [r21, r22, r23]
})

print("\nComparison of Models:")
print(results)

```

```python

# 10. APPLY LASSO AND RIDGE REGULARIZATION


from sklearn.linear_model import Lasso, Ridge
from sklearn .preprocessing import StandardScaler 
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# ---------- 10.1 LASSO REGRESSION ----------
lasso = Lasso(alpha=0.1)  # you can tune alpha
lasso.fit(X_train_scaled, Y_train)
Y_pred_lasso = lasso.predict(X_test_scaled)


mse_lasso = mean_squared_error(Y_test, Y_pred_lasso)
rmse_lasso = np.sqrt(mse_lasso)
r2_lasso = r2_score(Y_test, Y_pred_lasso)

print("\nLASSO REGRESSION RESULTS (alpha=0.1):")
lasso_coef_df = pd.DataFrame({
    "Feature": X.columns,
    "Coefficient": lasso.coef_
})
print(lasso_coef_df)
print("Number of coefficients shrunk to 0:", np.sum(lasso.coef_ == 0))

print("\nLasso Performance:")
print("MSE:",mse_lasso)
print("RMSE:", rmse_lasso)
print("R^2 :", r2_lasso)

# ---------- 10.2 RIDGE REGRESSION ----------
ridge = Ridge(alpha=1.0)  # you can tune alpha
ridge.fit(X_train_scaled, Y_train)
Y_pred_ridge = ridge.predict(X_test_scaled)


mse_ridge = mean_squared_error(Y_test, Y_pred_ridge)
rmse_ridge = np.sqrt(mse_ridge)
r2_ridge = r2_score(Y_test, Y_pred_ridge)

print("\nRIDGE REGRESSION RESULTS (alpha=1.0):")
ridge_coef_df = pd.DataFrame({
    "Feature": X.columns,
    "Coefficient": ridge.coef_
})
print(ridge_coef_df)

print("\nRidge Performance:")
print("MSE:",mse_ridge)
print("RMSE:", rmse_ridge)
print("R^2 :", r2_ridge)
```
