```python
import numpy as np
from scipy import stats
```

```python
# Given data
sample_mean = 3050        # x̄
mu_x = 600                # mean of X (units)
sigma_x = 25              # standard deviation of X
n = 25                    # sample size
alpha = 0.05              # significance level
```

```python
# Step 1: Theoretical mean weekly cost from model
mu_w = 1000 + 5 * mu_x    

# Step 2: Standard deviation of weekly cost (W = 1000 + 5X)
sigma_w = 5 * sigma_x  

se = sigma_w / np.sqrt(n)
```

```python
# Test statistic (z)
z_stat = (sample_mean - mu_w) / se
```

```python

# Critical value for right-tailed test at alpha = 0.05
z_critical = stats.norm.ppf(1 - alpha)
```

```python
# p-value for right-tailed test (μ > μ0)
p_value = 1 - stats.norm.cdf(z_stat)
```

```python
print("Theoretical mean weekly cost (μ):", mu_w)
print("Standard deviation of weekly cost (σ):", sigma_w)
print("Standard error (SE):", se)
print("Test statistic (z):", z_stat)
print("Critical value (z_alpha):", z_critical)
print("p-value:", p_value)

# Step 4: Decision
if z_stat > z_critical:
    print("\nDecision: Reject H0.")
else:
    print("\nDecision: Do NOT reject H0.")

# Step 5: Conclusion
print("\nConclusion:")
print("Based on the sample, there is not enough evidence to say that the")
print("weekly operating cost is higher than the model suggests.")
```
