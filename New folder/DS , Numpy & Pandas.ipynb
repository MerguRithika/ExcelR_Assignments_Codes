```python
#Section C: Exploring Pandas
#    11.Create a small DataFrame manually with 10 rows and columns: 'Name', 'Age', 'City', and 'Salary'. Then:
#- Select 'Name' and 'City' columns
#- Drop 'City' column
#- Fill any missing values in 'Salary' column with the mean
#- Remove any duplicate rows
data = {
    'Name': ['jaya', 'sai', 'rithu', 'virat', 'shilpa',
             'vinay', 'kumar', 'shree', 'chandana', 'Julie'],
    'Age': [25, 30, 35, 28, 40,
            32, 27, 29, 31, 26],
    'City': ['Hyd', 'Chennai', 'Delhi', 'Pune', 'Mysore',
             'Kolkata', 'Benguloor', 'Maharastra', 'Vizag', 'Hyd'],
    'Salary': [50000,np.nan, 75000, 52000, 82000,
               58000, 49000, 53000, np.nan, 56000]
}
df = pd.DataFrame(data)

print(df)
```
