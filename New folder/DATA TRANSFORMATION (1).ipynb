```python
df_countable = df[['age', 'fnlwgt', 'education_num', 'capital_gain', 'capital_loss', 'hours_per_week']]
df_countable.head()
```

```python
# standardization SCALING 
from sklearn .preprocessing import StandardScaler 
SS = StandardScaler()
SS_X = SS.fit_transform(df_countable)
SS_X = pd.DataFrame(SS_X)
SS_X.columns = list(df_countable)
SS_X.head()
```

```python
# Select categorical columns
cat_cols = df.select_dtypes(include='object').columns
print("Categorical Columns:", cat_cols)
```

```python
df
```

```python
# Columns with < 5 categories → One-Hot Encoding
onehot_cols = [col for col in cat_cols if df[col].nunique() < 5]

# Columns with >= 5 categories → Label Encoding
label_cols = [col for col in cat_cols if df[col].nunique() >= 5]

print("One-Hot Columns:", onehot_cols)
print("Label Encoding Columns:", label_cols)
```

```python
df = pd.get_dummies(df, columns=onehot_cols, drop_first=True)
df
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for col in label_cols:
    df[col] = le.fit_transform(df[col])
```

```python
print(df.head())
print(df.dtypes)
```

```python
import pandas as pd
import numpy as np

# Feature 1: Work Hours Category
df['work_hours_category'] = pd.cut(
    df['hours_per_week'],
    bins=[0, 30, 40, 60, 100],
    labels=['Part_time', 'Full_time', 'Over_time', 'Extreme']
)

# Feature 2: Capital Net
df['capital_net'] = df['capital_gain'] - df['capital_loss']

# Transformation: Log on capital_gain
df['capital_gain_log'] = np.log1p(df['capital_gain'])

# Quick check
print(df[['hours_per_week', 'work_hours_category', 'capital_gain', 'capital_loss', 'capital_net', 'capital_gain_log']].head())

```
